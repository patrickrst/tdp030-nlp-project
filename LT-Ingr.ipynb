```python
ingr = set()
ingrGroups = []
for x in data[:10000]:            #change the number in here for the amount of recipes you want to use
    ingrGroups.append([])
    for i in range(len(x['ingredients'])):
        if x['valid'][i]:
            ingr.add(x['ingredients'][i]['text'])
            ingrGroups[-1].append(x['ingredients'][i]['text'])
ingrSorted = sorted(ingr)
ingrSorted.remove('')
            
```

```python
#Finds the amount of times an ingredient occurs with another ingredient
probs = {}
cnt = collections.Counter()
for grp in ingrGroups:
    for ingr in grp:
        cnt[ingr] += 1

u = 0        
#for i in ingrSorted:
#    probs[i] = {}
#    for x in ingrSorted:
#        if i != x:
#            for grp in ingrGroups:
#                if x in grp and i in grp:
#                    if x in probs[i].keys():
#                        probs[i][x] = probs[i][x] + 1
#                    else:
#                        probs[i][x] = 1

def getProbs(ingredient):
    probs = {}
    for i in ingrSorted:
        if i != ingredient:
            for grp in ingrGroups:
                if ingredient in grp and i in grp:
                    if i in probs.keys():
                        probs[i] = probs[i] + 1
                    else:
                        probs[i] = 1
                #u += 1
                #b = str(u) + "/" + str(1*len(ingr)*len(ingrGroups))
                #print(b, end="\r")
    return probs
```

```python
#This function prints the random ingredients based on the main ingredient and returns a list of the ingredients.
#param main = The main ingredient for the recipe
#param nmbrOfIngr = The number of ingredients the recipe contains
def getIngr(main, nmbrOfIngr):
    ingrs = []
    ingrs.append(main)
    for n in range(nmbrOfIngr):
        ingr = main
        rand = random.randint(0, cnt[ingr])
        i = 0
        probs = getProbs(ingr)
        for key in probs.keys():
            i += probs[key]
            if i >= rand and key not in ingrs:
                print(key)
                ingr = key
                ingrs.append(key)
                break
    return ingrs

getIngr("Chinese cabbage", 6)
```
